```python
# Intersting number pattern generator

n = 1
m = 2

N = 100 # Number of numbers we are going to go through

def pattern_gen(n, m, nums):
    b = 0
    i = 0
    new_pattern = []
    for num in nums:
        if b == 0:
            if i >= n:
                i = 0
                b = 1
            else:
                new_pattern.append(num)
                i += 1
        else:
            if i >= m:
                i = 0
                b = 0
            else:
                new_pattern.append(num)
                i += 1
    return new_pattern

v = pattern_gen(7,2, list(range(1, 10000)))
    
```

```python
def cons_sum(nums):
    new_sum = []
    sum_ = 0
    for num in nums:
        sum_ += num
        new_sum.append(sum_)
    return new_sum
```

```python
test = cons_sum(v)
test
```

```python
test_2 = cons_sum(pattern_gen(2, 7, test))
test_2
```

```python
import matplotlib.pyplot as plt

plt.plot(range(len(test_2)), test_2)
```

```python

# Let's randomize the pattern 

import numpy as np

choice = list(range(1, 100))

m_c = np.random.choice(choice, 4, replace = False)
m_c
```

```python
rand_1 = cons_sum(pattern_gen(m_c[0], m_c[1], list(range(1, 10000))))
rand_2 = cons_sum(pattern_gen(m_c[2], m_c[3], rand_1))
rand_2
```

```python
plt.plot(range(len(rand_2)), rand_2)
plt.plot(range(len(test_2)), test_2)
```

```python
def rat_sum(nums):
    rat_nums = []
    sum_ = 0
    for num in nums:
        sum_ += num
        rat_nums.append(num/sum_)
    
    return rat_nums

rand_1_rat = rat_sum(pattern_gen(m_c[2], m_c[3], rand_1))
plt.plot(range(len(rand_1_rat)),rand_1_rat)
```

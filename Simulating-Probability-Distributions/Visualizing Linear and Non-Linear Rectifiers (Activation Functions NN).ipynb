```python
# Description Testing Visualizing Rectifiers (Activation Functions) 

import numpy as np
import pandas as pd
from scipy import stats as st
import matplotlib.pyplot as plt
```

```python
# We are going to generate some points to plot; N will denote the "size" of our closed interval
N = 10
df = pd.DataFrame({"x_values" :np.linspace(-N, N, num = 1001)})
```

```python
# Next we want to define a general rectifier methods
def _rectifier_LV_(a):
    return lambda x: x if x > 0 else max(x, x*a)
# List of possible constants
const = np.linspace(-5, 5, num = 11)

# Plot the variety of rectifier methods
plt.figure(figsize =(12.5, 4.5))

for i in range(0, len(const)):
    df["y_values"] = df["x_values"].apply(_rectifier_LV_(const[i]))
    plt.plot(df["x_values"], df["y_values"], label = "Line" +" " + "LV_Rect" + " " + str(const[i]))

plt.legend()
plt.show()
    
```

```python
# Let's plot some Non_linear Variants

def _GELU_(mean, sd):
    return lambda x: x*st.norm.cdf(x, loc = mean, scale = sd)

means = np.linspace(-5, 5, num = 11)
sd_s = np.linspace(-2, 2, num = 13)

plt.figure(figsize =(13, 13))
# We are going to plot the _GELU_'s 
for i in range(0, len(means)):
    for j in range(0, len(sd_s)):
        df["y_values"] = df["x_values"].apply(_GELU_(means[i], sd_s[j]))
        plt.plot(df["x_values"], df["y_values"], label = "GELU: Mean = " + str(means[i]) + ", SD = " + str(sd_s[j]))
plt.show()
```

```python
# Plot another variant: the sigmoid f(x) = 1 / (1+e^(-cx), c > 0

c_ = np.linspace(0, 4, num = 11)
# import math for the euler constant
import math 

def _SIGLU_(c):
    return lambda x: 1/(1+math.e**(-c*x))

# Plot the sigmoid

plt.figure(figsize =(13, 13))
for i in range(0, len(c_)):
    df["y_values"] = df["x_values"].apply(_SIGLU_(c_[i]))
    plt.plot(df["x_values"], df["y_values"], label = "Line" +" " + "SIGLU: const = " + " " + str(c_[i]))
plt.show()
```

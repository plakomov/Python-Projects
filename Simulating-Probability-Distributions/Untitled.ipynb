```python
# Prime Binary Number Grid

# Create a fixed size grid of Size N x N

# Step 1: Create a binary representation of size N of the natural numbers, where 1 represent a prime number 
# and 0 represents a composite number

N = 30

import numpy as np
import math

def check_prime(num):
    if num == 1: return False
    else: 
        i = 2
        while i <= math.sqrt(num):
            if num % i == 0: return False
            else:
                i += 1
    return True

def bin_prime(k, n):
    bin_pr = np.zeros(n-k+1)
    i = k 
    while i <= n:
        if check_prime(i):
            bin_pr[i-k] = 1
        i +=1 
    return bin_pr
        
    
```

```python
bin_prime(1,30)
```

```python
def bin_prime_matrix(n):
    matrix = bin_prime(1, n)
    k = 2
    while k <= n: 
        matrix = np.vstack((matrix, bin_prime(k, n + (k-1))))
        k += 1
    return matrix
bin_prime_matrix(10)
```

```python
N = 1000
def trace(sq_matrix):
    trace = 0
    for i in range(sq_matrix.shape[0]):
        trace += sq_matrix[i][i]
    return trace

traces = []

for i in range(2, N+1):
    traces.append(trace(bin_prime_matrix(i)))
    
traces
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 12))
plt.plot(list(range(2,N+1)), traces)
```

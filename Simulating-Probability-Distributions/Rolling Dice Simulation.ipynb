```python
# 1) We are going to simulate rolling N dices 


import numpy as np
import matplotlib.pyplot as plt
```

```python


# Simulate a roll of dice and getting 6 
# k>= 1
def roll_dices(k, m):
    arr = np.zeros(k)
    arr[0] = 1
    np.random.shuffle(arr)
    return np.random.choice(arr, size = int(m))
```

```python
# We design a simulation with a intial condition of the number of die to roll

def sim_roll_dice(n, p):
    lvl = 0
    dice_thr = 0
    num_dice = n
    while num_dice != 0:
        dice_thr += num_dice
        rolled = roll_dices(p, num_dice)
        num_dice = num_dice - np.sum(rolled)
        lvl += 1
    return lvl, dice_thr
```

```python
sim_roll_dice(5, 6)
```

```python
n_epoch = 10000
N = 5
p = 6
dice_rolled = []
lvls = []

for i in range(n_epoch):
    lvl, dice_ = sim_roll_dice(N, p)
    dice_rolled.append(dice_)
    lvls.append(lvl)

dice_rolled
```

```python
plt.figure(figsize = (12.5, 12.5))
plt.hist(dice_rolled, int(max(dice_rolled)-min(dice_rolled)), (min(dice_rolled) -0.5, max(dice_rolled) + 0.5))
```

```python
plt.figure(figsize = (12.5, 12.5))
plt.hist(lvls, int(max(lvls)-min(lvls)), (min(lvls) -0.5, max(lvls) + 0.5))
```
